Importing Library

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```
dataset = pd.read_csv('/content/mushrooms.csv')
```

```
dataset.head(10)
```

```
dataset.info()
```

```
dataset.describe()
```

```
for column in dataset.columns:
  print(column,"Unique Values:",dataset[column].unique())
  print("---"*30)
```

```
columns = dataset.columns
```

Encoding the categorical features

```
X = dataset.drop('class', axis = 1)
y = dataset['class']
```

```
X
```

```
for column in X:
  dummy = pd.get_dummies(dataset[column], drop_first = True)
  dataset.drop(column, axis = 1, inplace = True)
  dataset = pd.concat([dataset, dummy], axis = 1)
```

```
dataset.head()
```

```
dataset.drop('class', axis = 1, inplace = True)
```

```
dataset.head()
```

```
dummy = pd.get_dummies(y, drop_first = True)
```

```
y = pd.concat([y, dummy], axis = 1)
```

```
y
```

```
y.drop('class', inplace = True, axis = 1)
```

```
y
```

```
X = dataset
```

```
X.isna().sum()
```

Splitting data into train and test

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 355)
```

```
X_train.shape
```

Normalizing the data

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```
X_train = sc.fit_transform(X_train)
```

```
X_test = sc.transform(X_test)
```

```
from sklearn.linear_model import SGDClassifier
sgd = SGDClassifier()
```

```
sgd.fit(X_train, y_train)
```

```
y_pred = sgd.predict(X_test)
```

```
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
score = accuracy_score(y_test, y_pred)
```
